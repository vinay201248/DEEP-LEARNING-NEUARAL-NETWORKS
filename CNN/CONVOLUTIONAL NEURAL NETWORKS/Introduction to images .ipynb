```python
# checking pixel intensity colors using matplotlib 
import numpy as np

values = np.array([0,5,7,11,15])
print(values)
print(values.shape)
values = values.reshape(1 , -1)
print(values.shape)
print('----------------')
plt.imshow(values,  cmap = 'gray')
```

```python
# checking pixel intensity colors using matplotlib 
import numpy as np

values = np.array([-1,50,7,111,15])
print(values)
print(values.shape)
values = values.reshape(1 , -1)
print(values.shape)
print('----------------')
plt.imshow(values,  cmap = 'gray')
```

```python
# checking how cpu takes color using pixel from 0 to 255


a = np.arange(0,256)
a = a.reshape(1,-1)
a = np.repeat(a , 100 , axis = 0)
plt.imshow(a,cmap = 'gray')
```

```python
# checking for each channel 

plt.subplot(2,2,1)
plt.title('R_channel')
plt.imshow(color_image[: , : , 0] , cmap = 'gray')  # R channel

plt.subplot(2,2,2)
plt.title('G_channel')
plt.imshow(color_image[: , : , 1],cmap = 'gray')  # g channel

plt.subplot(2,2,3)
plt.title('B_channel')
plt.imshow(color_image[: , : , 2],cmap = 'gray')  # b channel

plt.subplot(2,2,4)
plt.title('original_image')
plt.imshow(color_image[: , : , :]) 

```
