```python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn
import tensorflow as tf
```

```python
df = pd.read_csv('./Car_Purchasing_Data.csv',encoding='ISO-8859-1')
```

```python
df
```

```python
df.columns
```

```python
df = df.drop(['Customer Name', 'Customer e-mail', 'Country'],axis = 1)
```

```python
df
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
x = df.drop(['Car Purchase Amount'],axis = 1)
```

```python
y = df['Car Purchase Amount']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.33, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
x1 = x_train[: , 1]
```

```python
y = y.values.reshape(-1,1)
y_train_scaled = sc.fit_transform(y)
y_train_scaled
```

```python
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Activation
from tensorflow.keras.activations import sigmoid,relu,softmax
```

```python
x1.shape
```

```python
# Architecture 

model = Sequential()

# First Hidden layer 

model.add(Dense(units = 256,activation='relu',input_dim = 1))

# second layer 

model.add(Dense(units=128,activation='relu',))


#Third layer 

model.add(Dense(units=64 ,activation='relu',))


# fourth Layer 

model.add(Dense(units=32 ,activation='relu'))

# fifth Layer 

model.add(Dense(units=16 ,activation='relu',))


# sixth Layer 

model.add(Dense(units=8 ,activation='relu'))


# Seventh layer 

model.add(Dense(units=4,activation='relu'))


# Eith layer 

model.add(Dense(units= 1 ,activation='linear'))



```

```python
model.summary()
```

```python
model.compile(optimizer='adam',loss='mean_squared_error',metrics=['acc'])
```

```python
#    

reg = model.fit(x1 , y_train_scaled, epochs= 50 ,validation_split=0.2,batch_size = 20)
```

```python
model.history.history.keys()
```

```python
model.history.history['loss']
```

```python
model.history.history['accuracy']
```

```python
plt.subplot(1,2,1)
plt.plot(np.arange(50) , model.history.history['loss'] , color = 'r' , label = 'train_loss')
plt.plot(np.arange(50) , model.history.history['accuracy'] , color = 'b' , label = 'train_acc')
plt.title('training_part')

plt.subplot(1,2,2)
plt.plot(np.arange(50) , model.history.history['val_loss'] , color = 'r' , label = 'val_loss')
plt.plot(np.arange(50) , model.history.history['val_accuracy'] , color = 'b' , label = 'val_acc')
plt.title('validation_part')
plt.legend(loc = 0)
plt.show()

```
